Write a program that reads something and shows on the screen its primitive type and all possible information about it.

```python
n = input('Type something: ')
print('The primitive type of this value is', type(n))
print('Are all characters spaces?', n.isspace())
print('Is it a number?', n.isnumeric())
print('Is it alphabetical?', n.isalpha())
print('Is it alphanumeric?', n.isalnum())
print('Is it in uppercase?', n.isupper())
print('Is it in lower case?', n.islower())
print('Does each word start with an upper case letter?', n.istitle())
```
