Write a program that reads an integer and shows its predecessor and successor.

```python
n1 = int(input('Enter a number: '))
print('The predecessor of {} is {}.'.format(n1, n1-1))
print('The successor of {} is {}.'.format(n1, n1+1))
```

```python
n = int(input('Enter another number: '))
a = n-1
s = n+1
print('The predecessor of {} is {} and his successor is {}.'.format(n, a, s))
```
